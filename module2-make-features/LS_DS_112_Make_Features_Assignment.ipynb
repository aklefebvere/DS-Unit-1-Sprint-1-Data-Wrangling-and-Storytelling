```python
#We need to remove the extra NaN's its creating
df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2, engine='python')
df.isnull().sum().sort_values()
```

```python
#Making sure that the term column doesn't have any NaN values that could mess up the int conversion
df['term'].isnull().sum()
```

```python
#function to take a variable and strip the spaces and the word "months"
def str_to_int (cell_contents):
  cell_contents = cell_contents.strip().strip('months').strip()
  return int(cell_contents)


#apply the str_to_int function to every cell of the term column
df['term'] = df['term'].apply(str_to_int)




```

```python
#checking if term are ints and was stripped correctly
df['term'].head(5)
```

```python
#created a empty list to be used to fill 1's and 0's in 
test = []
#for statement to check if the loan status is "Current" or "Fully Paid" and give it a 1 or 0
for status in df['loan_status']:
  if status == "Current" or status == "Fully Paid":
    test.append(1)
  else:
    test.append(0)
#created a column called "loan_status_is_great" and gave it the values of list
df['loan_status_is_great'] = test
df.head()
```

```python
#The NaN values need to be replaced before its converted into datatime64
df['last_pymnt_d'].isnull().sum()
```

```python
#Now that there are no NaN values, we don't have to worry about float values in the conversion
df['last_pymnt_d'] = df['last_pymnt_d'].fillna(0)
df['last_pymnt_d'].isnull().sum()
```

```python
#convert the last_pymnt_d into datetime64 dtype so that it can be used to create other time columns
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)
```

```python
#Checking if the datatype converted correctly
#<M8[ns] == datatime64[ns]
df['last_pymnt_d'].dtypes
```

```python
#creating last_pymnt_d_month column
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month
```

```python
#double checking that there are no NaN values
df['last_pymnt_d_month'].isnull().sum()
```

```python
#creating last_pymnt_d_year column
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year
```
